# Uninformed Search Algorithms

Breadth-First Search (BFS)

```python
from collections import deque

# BFS for single connected component
def bfs(adj):
    V = len(adj)
    visited = [False] * V
    res = []

    src = 0
    q = deque()
    visited[src] = True
    q.append(src)

    while q:
        curr = q.popleft()
        res.append(curr)

        # visit all the unvisited
        # neighbours of current node
        for x in adj[curr]:
            if not visited[x]:
                visited[x] = True
                q.append(x)

    return res

def addEdge(adj, u, v):
    adj[u].append(v)
    adj[v].append(u)


if __name__ == "__main__":
    V = 5
    adj = []

    # creating adjacency list
    for i in range(V):
        adj.append([])

    addEdge(adj, 1, 2)
    addEdge(adj, 1, 0)
    addEdge(adj, 2, 0)
    addEdge(adj, 2, 3)
    addEdge(adj, 2, 4)

    res = bfs(adj)

    for node in res:
        print(node, end=" ")
```

Depth-First Search (DFS)

```python
def add_edge(adj, s, t):
    # Add edge from vertex s to t
    adj[s].append(t)

def dfs_rec(adj, visited, s):
    # Mark the current vertex as visited
    visited[s] = True

    # Print the current vertex
    print(s, end=" ")

    # Recursively visit all adjacent vertices
    # that are not visited yet
    for i in adj[s]:
        if not visited[i]:
            dfs_rec(adj, visited, i)


def dfs(adj, s):
    visited = [False] * len(adj)
    # Call the recursive DFS function
    dfs_rec(adj, visited, s)


if __name__ == "__main__":
    V = 5

    # Create an adjacency list for the graph
    adj = [[] for _ in range(V)]

    # Define the edges of the graph
    edges = [[1, 2], [1, 0], [2, 0], [2, 3], [2, 4]]

    # Populate the adjacency list with edges
    for e in edges:
        add_edge(adj, e[0], e[1])

    source = 1
    print("DFS from source:", source)
    dfs(adj, source)
```

Iterative Deepening Search (IDS)

```python
from collections import defaultdict


# -----------------------------------
# STEP 1: Graph Class Definition
# -----------------------------------
class Graph:

    def __init__(self, vertices):
        self.V = vertices
        self.graph = defaultdict(list)

    # Add directed edge
    def addEdge(self, u, v):
        self.graph[u].append(v)

    # -----------------------------------
    # STEP 2: Depth-Limited Search (DLS)
    # -----------------------------------
    def DLS(self, current, target, depth, path):

        print(f"Visiting: {current}, Depth remaining: {depth}")

        path.append(current)

        # Goal Test
        if current == target:
            return True

        # Depth limit reached
        if depth == 0:
            path.pop()
            return False

        # Explore neighbors
        for neighbor in self.graph[current]:
            if self.DLS(neighbor, target, depth - 1, path):
                return True

        path.pop()
        return False

    # -----------------------------------
    # STEP 3: Iterative Deepening Search
    # -----------------------------------
    def IDDFS(self, src, target, maxDepth):

        for depth in range(maxDepth + 1):
            print("\n==========================")
            print(f"Searching with Depth Limit = {depth}")
            print("==========================")

            path = []

            if self.DLS(src, target, depth, path):
                print("\nTarget Found!")
                print("Path:", " -> ".join(map(str, path)))
                print("Depth Level:", depth)
                return True

        print("\nTarget NOT reachable within max depth.")
        return False


# -----------------------------------
# STEP 4: Create Graph
# -----------------------------------
g = Graph(7)

g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 3)
g.addEdge(1, 4)
g.addEdge(2, 5)
g.addEdge(2, 6)

# -----------------------------------
# STEP 5: Run IDS
# -----------------------------------
src = 0
target = 6
maxDepth = 3

g.IDDFS(src, target, maxDepth)
```

# Informed Search Algorithms:

Simple Hill Climbing

```python
# ------------------------------
# STEP 1: Define Initial & Goal
# ------------------------------
board = [[1,4,7],
         [2,5,8],
         [3,6,0]]

goal = [[1,2,4],
        [5,0,7],
        [3,6,8]]


# ------------------------------
# STEP 2: Heuristic Function
# ------------------------------
# h(n) = number of misplaced tiles
def heuristic(board, goal):
    misplaced = 0
    for i in range(3):
        for j in range(3):
            if board[i][j] != 0 and board[i][j] != goal[i][j]:
                misplaced += 1
    return misplaced


# ------------------------------
# STEP 3: Find Blank Space
# ------------------------------
def blank_space(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j


# ------------------------------
# STEP 4: Print Board
# ------------------------------
def print_board(board):
    for row in board:
        print(row)
    print()


# ------------------------------
# STEP 5: Simple Hill Climbing
# ------------------------------
def hill_climbing(board, goal):
    x, y = blank_space(board)
    h_n = heuristic(board, goal)
    step = 0

    print("Initial Board:")
    print_board(board)
    print("Initial Heuristic:", h_n)
    print("---------------------")

    # Possible moves (Up, Down, Left, Right)
    moves = [(-1,0),(1,0),(0,-1),(0,1)]

    while True:

        if h_n == 0:
            print("Goal Reached!")
            return step

        moved = False

        for dx, dy in moves:
            n_x = x + dx
            n_y = y + dy

            if 0 <= n_x < 3 and 0 <= n_y < 3:

                # Swap blank with neighbor
                board[x][y], board[n_x][n_y] = board[n_x][n_y], board[x][y]

                h_new = heuristic(board, goal)

                print("Trying Move:")
                print_board(board)
                print("Heuristic:", h_new)

                # Accept move only if better
                if h_new < h_n:
                    x, y = n_x, n_y
                    h_n = h_new
                    step += 1
                    moved = True
                    print("Move Accepted | Step:", step)
                    print("---------------------")
                    break
                else:
                    # Revert move
                    board[x][y], board[n_x][n_y] = board[n_x][n_y], board[x][y]

        if not moved:
            print("Stuck in Local Optima!")
            print("Final Board:")
            print_board(board)
            return step


# ------------------------------
# STEP 6: Run Algorithm
# ------------------------------
total_steps = hill_climbing(board, goal)
print("Total Steps:", total_steps)
```

Steepest Ascent Hill Climbing

```python
board = [[1,4,7],
         [2,5,8],
         [3,6,0]]

goal = [[1,2,4],
        [5,0,7],
        [3,6,8]]

def heuristic(board, goal):
    misplaced = 0
    for i in range(3):
        for j in range(3):
            if board[i][j] != 0 and board[i][j] != goal[i][j]:
                misplaced += 1
    return misplaced


def blank_space(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j


def print_board(board):
    for row in board:
        print(row)
    print()

def steepest_hill_climbing(board, goal):
    x, y = blank_space(board)
    h_n = heuristic(board, goal)
    step = 0

    print("Initial State:")
    print_board(board)
    print("Initial Heuristic:", h_n)
    print("----------------------")

    moves = [(-1,0), (1,0), (0,-1), (0,1)]

    while True:

        if h_n == 0:
            print("Goal Reached!")
            return step

        h_min = h_n
        best_move = None

        # Evaluate ALL neighbors
        for dx, dy in moves:
            n_x = x + dx
            n_y = y + dy

            if 0 <= n_x < 3 and 0 <= n_y < 3:

                # Try move
                board[x][y], board[n_x][n_y] = board[n_x][n_y], board[x][y]
                h_new = heuristic(board, goal)

                print("Checking Move:")
                print_board(board)
                print("Heuristic:", h_new)

                # Undo move
                board[x][y], board[n_x][n_y] = board[n_x][n_y], board[x][y]

                if h_new < h_min:
                    h_min = h_new
                    best_move = (n_x, n_y)

        # If no better neighbor
        if best_move is None:
            print("Stuck in Local Optima / Plateau")
            print("Final State:")
            print_board(board)
            return step

        # Apply best move
        n_x, n_y = best_move
        board[x][y], board[n_x][n_y] = board[n_x][n_y], board[x][y]
        x, y = n_x, n_y
        h_n = h_min
        step += 1

        print("Best Move Applied")
        print("Step:", step)
        print("New Heuristic:", h_n)
        print_board(board)
        print("----------------------")


total_steps = steepest_hill_climbing(board, goal)
print("Total Steps:", total_steps)
```

```python
from queue import PriorityQueue

# ------------------------------
# STEP 1: Input Graph
# ------------------------------
edge = int(input("Enter the number of edges: "))
graph = {}

for i in range(edge):
    u, v, w = input("Enter edge (u v w): ").split()
    w = int(w)

    if u not in graph:
        graph[u] = []
    if v not in graph:
        graph[v] = []

    graph[u].append((v, w))
    graph[v].append((u, w))   # remove this line if graph is directed

print("\nGraph:", graph)


# ------------------------------
# STEP 2: Input Heuristic Values
# ------------------------------
city = int(input("\nEnter number of cities (heuristic values): "))
heuristic = {}

for i in range(city):
    node, h = input("Enter city and heuristic value: ").split()
    heuristic[node] = int(h)

print("\nHeuristic:", heuristic)


# ------------------------------
# STEP 3: Greedy Best First Search
# ------------------------------
def greedy_bfs(start, goal, graph, heuristic):
    pq = PriorityQueue()
    visited = set()

    # push (heuristic, node, path, cost)
    pq.put((heuristic[start], start, [start], 0))

    while not pq.empty():
        h, current, path, cost = pq.get()

        if current in visited:
            continue

        print("Expanding:", current)
        visited.add(current)

        if current == goal:
            print("\nGoal reached!")
            print("Path:", " -> ".join(path))
            print("Total Cost:", cost)
            return path, cost

        for neighbor, weight in graph[current]:
            if neighbor not in visited:
                pq.put((
                    heuristic[neighbor],
                    neighbor,
                    path + [neighbor],
                    cost + weight
                ))

    print("No path found.")
    return None, None


# ------------------------------
# STEP 4: Take Start & Goal
# ------------------------------
start_node = input("\nEnter start node: ")
goal_node = input("Enter goal node: ")

greedy_bfs(start_node, goal_node, graph, heuristic)
```

A* Algorithm

```python
from queue import PriorityQueue
import itertools # Import itertools

# -------------------- Initial & Goal --------------------

start =start = [
 [1,4,7],
 [2,5,8],
 [3,6,0]
]


goal = [
 [1,2,3],
 [4,5,6],
 [8,7,0]
]



# -------------------- Utilities --------------------

def to_tuple(state):
    return tuple(tuple(row) for row in state)

def print_state(state):
    for row in state:
        print(row)
    print()

def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

# -------------------- Manhattan Heuristic --------------------

def manhattan(state, goal):
    dist = 0
    for i in range(3):
        for j in range(3):
            val = state[i][j]
            if val != 0:
                for x in range(3):
                    for y in range(3):
                        if goal[x][y] == val:
                            dist += abs(i - x) + abs(j - y)
    return dist

# -------------------- Neighbors --------------------

def get_neighbors(state):
    x, y = find_blank(state)
    neighbors = []

    for dx, dy, mv in [(-1,0,'U'), (1,0,'D'), (0,-1,'L'), (0,1,'R')]:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append((new_state, mv))

    return neighbors

# -------------------- A* SEARCH --------------------

def A_star(start, goal):
    pq = PriorityQueue()
    visited = set()
    g_cost = {}
    parent = {}

    counter = itertools.count() # Add a counter for tie-breaking

    start_key = to_tuple(start)
    goal_key = to_tuple(goal)

    g_cost[start_key] = 0
    parent[start_key] = (None, None)

    h0 = manhattan(start, goal)
    pq.put((h0, 0, next(counter), start_key, start))   # (f, g, count, key, state)

    while not pq.empty():
        f, g, count, cur_key, cur_state = pq.get() # Get count as well

        if cur_key in visited:
            continue
        visited.add(cur_key)

        if cur_key == goal_key:
            return parent, cur_key

        for nxt_state, mv in get_neighbors(cur_state):
            nxt_key = to_tuple(nxt_state)
            new_g = g + 1

            if nxt_key not in g_cost or new_g < g_cost[nxt_key]:
                g_cost[nxt_key] = new_g
                new_f = new_g + manhattan(nxt_state, goal)
                pq.put((new_f, new_g, next(counter), nxt_key, nxt_state)) # Put count
                parent[nxt_key] = (cur_key, mv)

    return None, None

# -------------------- Reconstruct Path --------------------

def reconstruct(parent, end_key, start):
    path = []
    moves = []

    cur = end_key
    while parent[cur][0] is not None:
        prev, mv = parent[cur]
        path.append(cur)
        moves.append(mv)
        cur = prev

    path.append(to_tuple(start))
    path.reverse()
    moves.reverse()
    return path, moves

# -------------------- RUN --------------------

parent, end_key = A_star(start, goal)

if parent is None:
    print("No solution found.")
else:
    path, moves = reconstruct(parent, end_key, start)
    print("Moves:", "".join(moves))
    print("Total steps:", len(moves))
    print("\nSolution path:")
    for state in path:
        print_state([list(row) for row in state])
```

# Evolutionary Algorithm

Genetic Algorithms

```python
import random

# Genetic Algorithm parameters
POPULATION_SIZE = 50
MUTATION_RATE = 0.1
MAX_GENERATIONS = 100

# Generate a random board state
def generate_board_state():
    board_state = [random.randint(0, 7) for _ in range(8)]
    return board_state

# Calculate the fitness of a board state
def calculate_fitness(board_state):
    conflicts = 0
    for i in range(8):
        for j in range(i + 1, 8):
            if board_state[i] == board_state[j] or abs(board_state[i] - board_state[j]) == j - i:
                conflicts += 1
    return 28 - conflicts  # Max fitness = 28 (no conflicts)

# Select parents for crossover using tournament selection
def tournament_selection(population):
    tournament_size = 5
    tournament = random.sample(population, tournament_size)
    return max(tournament, key=lambda x: x[1])

# Crossover operation (single-point crossover)
def crossover(parent1, parent2):
    crossover_point = random.randint(1, 7)
    child = parent1[:crossover_point] + parent2[crossover_point:]
    return child

# Mutation operation (swap two positions)
def mutate(board_state):
    pos1, pos2 = random.sample(range(8), 2)
    board_state[pos1], board_state[pos2] = board_state[pos2], board_state[pos1]
    return board_state

# Generate the initial population
population = [(generate_board_state(), 0) for _ in range(POPULATION_SIZE)]

# Main Genetic Algorithm loop
for generation in range(MAX_GENERATIONS):
    # Calculate fitness for each board state
    population = [(board_state, calculate_fitness(board_state)) for board_state, _ in population]

    # Check if solution is found
    best_board_state = max(population, key=lambda x: x[1])[0]
    if calculate_fitness(best_board_state) == 28:
        print("Solution found in generation", generation)
        break

    # Create the next generation
    new_population = []

    # Elitism: Keep the best board state from the previous generation
    new_population.append(max(population, key=lambda x: x[1]))

    # Perform selection, crossover, and mutation
    while len(new_population) < POPULATION_SIZE:
        parent1 = tournament_selection(population)
        parent2 = tournament_selection(population)
        child = crossover(parent1[0], parent2[0])
        if random.random() < MUTATION_RATE:
            child = mutate(child)
        new_population.append((child, 0))

    # Update the population
    population = new_population

# Print the best solution
print("Best solution:", best_board_state)
```
